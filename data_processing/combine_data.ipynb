```python
import pandas as pd

consumption = pd.read_csv("./data/consumption.csv", index_col=0)
emission = pd.read_csv("./data/emission.csv", index_col=0)
```

```python
df = pd.concat([consumption, emission], axis=1)
df
```

```python
df = df.rename(
    columns={
        "1 Renewable energy (%)": "RenewableConsumptionShare",
        "1.1 Hydro power (%)": "HydroPowerConsumptionShare",
        "1.2 Wind power (%)": "WindPowerConsumptionShare",
        "1.3 Wood fuels (%)": "WoodFuelsConsumptionShare",
        "1.4 Other renewable energy (%)": "OtherRenewablesConsumptionShare",
        "2 Fossil fuels and peat (%)": "FossilFuelsPeatConsumptionShare",
        "2.1 Oil (fossil) (%)": "OilConsumptionShare",
        "2.2 Coal (%)": "CoalConsumptionShare",
        "2.3 Natural gas (%)": "NaturalGasConsumptionShare",
        "2.4 Peat (%)": "PeatConsumptionShare",
        "2.5 Other fossil fuels (%)": "OtherFossilsConsumptionShare",
        "3 Nuclear energy (%)": "NuclearEnergyConsumptionShare",
        "4 Net imports of electricity (%)": "NetImportsConsumptionShare",
        "5 Others (%)": "OthersConsumptionShare",
        "Share of renewable electricity generation (%)": "RenewableProductionShare",
        "CO2 emission coefficient for electricity generation (g CO2/kWh)": "CO2Emission",
        "CO2-equivalent emission coefficient for electricity generation (g CO2 eq./kWh)": "CO2EquivEmission",
        "Life cycle emission coefficient for electricity generation, (g CO2 eq./kWh)": "LifeCycleEmission",
    }
)
```

```python
df.to_csv("./data/combined.csv")
```

```python
df = df[["RenewableProductionShare", "LifeCycleEmission"]]
df.to_csv("./data/lin_reg_data.csv")
```
