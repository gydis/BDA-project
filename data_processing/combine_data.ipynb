```python
import pandas as pd

consumption = pd.read_csv("./data/consumption.csv", index_col=0)
production = pd.read_csv("./data/production.csv", index_col=0)
emission = pd.read_csv("./data/emission.csv", index_col=0)
```

```python
df = pd.concat([consumption, production, emission], axis=1)
df = df.dropna()
df
```

```python
df.to_csv("./data/combined.csv")
```

```python
production_emission_cols = [col for col in df.columns if "Production" in col or col == "CO2EquivEmission"]
production_emission_df = df[production_emission_cols]
production_emission_df.to_csv("./data/data.csv")
```
